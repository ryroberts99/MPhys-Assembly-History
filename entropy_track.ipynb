```python
def time_from_redshift(z, omega0, omega_lambda, h):
    if omega_lambda != 0 and np.abs(omega0 - 1.0 + omega_lambda) > 10**-6:
        print(f"Omega0 = {omega0}")
        print("Omega_tot = {}".format(omega0 + omega_lambda))
        print("Descrepency = {}".format(omega0 - 1 + omega_lambda))
        raise ValueError("Invalid inputs - see output for details.")

    hTime = 9.778 / h# Gyrs

    if omega_lambda == 0:
        if omega0 == 1:
            time = (2 / 3) * hTime / ((1 + z)**1.5)
        else:
            if omega0 > 1:
                omega0_z = omega0 * z
                factor1 = omega0 / (2 * (omega0 - 1)**1.5)
                term1 = np.arccos((omega0_z - omega0 + 2) / (omega0_z + omega0))
                term2 = 2 * np.sqrt((omega0 - 1) * (omega0_z + 1) / (omega0_z + omega0))
                factor2 = term1 - term2
                time = factor1 * factor2 * hTime

            else:
                raise ValueError("Failed tryint to intergrate for open, matter dominated cosmology.")

    else:
        factor1 = (2 / 3) / np.sqrt(omega_lambda)
        term1 = np.sqrt(omega_lambda / omega0) / ((1 + z)**1.5)
        term2 = np.sqrt(1 + (omega_lambda / (omega0 * (1 + z)**3)))
        factor2 = np.log(term1 + term2)
        time = factor1 * factor2 * hTime

    return time
```

```python
def s_30_value(halo, subhalo, tag, sim):
    first_subhalo = catalogue_read("FOF", "FirstSubhaloID", data_location, tag=tag, sim=sim, cgs_units=True)[halo-1]
    subhalo_index = first_subhalo + subhalo 
    m_gas = catalogue_read("Subhalo", "ApertureMeasurements/Mass/030kpc", data_location, tag=tag, sim=sim, cgs_units=True)[subhalo_index][0]
    m_dm = catalogue_read("Subhalo", "ApertureMeasurements/Mass/030kpc", data_location, tag=tag, sim=sim, cgs_units=True)[subhalo_index][1]
    m_star = catalogue_read("Subhalo", "ApertureMeasurements/Mass/030kpc", data_location, tag=tag, sim=sim, cgs_units=True)[subhalo_index][4]
    m_bh = catalogue_read("Subhalo", "ApertureMeasurements/Mass/030kpc", data_location, tag=tag, sim=sim, cgs_units=True)[subhalo_index][5]
    r_30 = 0.03
    m_30 = m_gas+m_dm+m_star+m_bh
    v_30 = np.sqrt((grav_const * m_30) / r_30)
    
    
    gas_temps = np.empty(shape = (3, 3))
    gas_nh = np.empty(shape = (3, 3))
    all_coords = np.empty(shape = (3, 3))
    
    filenum = 0
    snapfile = os.path.join(data_location, model, sim, f'snapshot_{tag}', f'snap_{tag}.{filenum}.hdf5')
    snapshot = pyread_eagle.EagleSnapshot(snapfile)

    boxsize = (snapshot.boxsize)/h
    expansion_factor = 1/(1+float("{}.{}".format(tag[5:8], tag[9:12])))
#    print(expansion_factor)

    centre = catalogue_read("Subhalo", "CentreOfPotential", data_location, tag=tag, sim=sim)[subhalo_index]
    
    for k in range(0, 3):
        centre[k] = centre[k]*expansion_factor
        
#    for j in range(0, 16):
#    filenum = j

    gas_temps = particle_read(0, "Temperature", lower_bounds, upper_bounds, data_location, snapshot, centre = centre, model = model, sim =sim, tag = tag, phys_units = False, cgs_units = True)
    gas_nh = particle_read(0, "Density", lower_bounds, upper_bounds, data_location, snapshot, centre = centre, model = model, sim =sim, tag = tag, phys_units = False, cgs_units = True)
    gas_nh /= (proton_mass/0.76)
    
    coords = particle_read(0, "Coordinates", lower_bounds, upper_bounds, data_location, snapshot, centre = centre, model = model, sim = sim, tag = tag)
    coords_centre_adjusted = coords - centre
    coords_box_wrapped = ((coords_centre_adjusted + (boxsize/2)) % boxsize) - (boxsize/2)
    all_coords = np.concatenate((all_coords, coords_box_wrapped))
    for x in range(0, 3):
        all_coords = all_coords[np.where(np.abs(all_coords[:,x]) > 1e-14)] 
    
    
    sfr = particle_read(0, "StarFormationRate", lower_bounds, upper_bounds, data_location, snapshot, centre = centre, model = model, sim = sim, tag = tag, phys_units = False, cgs_units = True)
    particle_mass = particle_read(0, "Mass", lower_bounds, upper_bounds, data_location, snapshot, centre = centre, model = model, sim = sim, tag = tag, phys_units = False, cgs_units = True)

    galaxy_centre = centre
    
    entropy = np.ones(len(sfr))
    entropy = gas_temps/(gas_nh**(2/3))
    entropy = k_boltzmann*entropy
    entropy = entropy/ev_to_erg

    radial_dist = np.sqrt((galaxy_centre[0] - coords[:,0])**2 + (galaxy_centre[1] - coords[:,1])**2 + (galaxy_centre[2] - coords[:,2])**2)
    radial_dist = [x / 1000 for x in radial_dist]
    r_30 = np.ones(len(radial_dist))
    r_30 = [x*(30/1000) for x in r_30]
    count = 0

    mask = np.where((sfr==0)&(gas_nh < 0.1)&(radial_dist < r_30))
    particle_mass = particle_mass[mask]
    entropy_plot = entropy[mask]
#    radial_plot = radial_dist[mask] 
    s_30 = 0
    sum_weight = np.sum(particle_mass)
    sum_weighted_entropy = entropy_plot*particle_mass
    sum_weighted_entropy = np.sum(sum_weighted_entropy)
    s_30 = sum_weighted_entropy/sum_weight
    
    return s_30
    
def create_s_30_graph(func):
    s_30_plot = []
    snapshot_redshift = []
    lookback_time = []

    for i, simulation in enumerate((assembly_history_early, assembly_history_organic, assembly_history_late)):
        
        s_30_plot.append(np.empty(shape = (len(tags), )))
        snapshot_redshift.append(np.array([float("{}.{}".format(tag[5:8], tag[9:12]))for tag in tags]))
        lookback_time.append(np.empty(shape = (len(tags), )))
        
        for j, tag in enumerate(tags):
            try:
                halo = simulation[tag]["halo"]
                subhalo = simulation[tag]["subhalo"]

                if halo is None:
                    raise LookupError("No history avalible.")
                
                
                s_30_plot[i][j] = func(halo, subhalo, tag, sim[i])
                z = float("{}.{}".format(tag[5:8], tag[9:12]))
                lookback_time[i][j] = time_from_redshift(z, omega_0, omega_lambda, h)
                 
            except LookupError:
                s_30_plot[i][j] = None  
                

        array_filter = True ^ np.isnan(s_30_plot[i])
        s_30_plot[i] = s_30_plot[i][array_filter]
        snapshot_redshift[i] = snapshot_redshift[i][array_filter] 
        lookback_time[i] = lookback_time[i][array_filter]
        

    fig, ax = plt.subplots(figsize=(10,10))
    ax2 = ax.twiny() 
    ax.plot(lookback_time[0], s_30_plot[0], label = "GM-Early", c='#33bbee')
    ax.plot(lookback_time[1], s_30_plot[1], label = "Organic", c='#ee3377')
    ax.plot(lookback_time[2], s_30_plot[2], label = "GM-Late", c='#ee7733')
    ax.set_xlabel("Time (Gyr)")
    ax.set_ylabel("$S_{30}$ (KeV  $cm^{-2}$)")
#    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.tick_params(which='both', direction="in") 
    ax2.set_xlabel('$z$')  # we already handled the x-label with ax1 
    ax2.tick_params(axis='x')
    ax2.set_xscale('log')
    ax2.set_yscale('log') 
    ax2.tick_params(which='both', direction="in")
    ax2.set_xlim(1e1, 1e-1)
    ax2.yaxis.tick_right()
    ax2.xaxis.set_label_position('top') 
    ax2.yaxis.set_label_position('right') 
    ax2.get_yaxis().set_visible(False)
    ax.set_title("$S_{30}$ Over Time for Each Halo Type")
    ax.yaxis.label.set_size(32)
    ax.xaxis.label.set_size(32)
    ax2.xaxis.label.set_size(32)
    ax.legend(loc='upper left',prop={'size': 18})

    plt.savefig('S_30.png')
    plt.show()
    
create_s_30_graph(s_30_value)
```
